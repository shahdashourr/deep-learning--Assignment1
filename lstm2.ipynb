```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

inputs = [1, 2, 3, 4]

h = 0
c = 0

wf, wi, wc, wo = 0.4, 0.6, 0.8, 0.5
bf, bi, bc, bo = 0.1, 0.1, 0.1, 0.1

print("LSTM Forward Pass\n")

for step, x in enumerate(inputs):

    
    forget_gate = sigmoid(wf * x + bf)

   
    input_gate = sigmoid(wi * x + bi)

 
    candidate = tanh(wc * x + bc)

  
    c = forget_gate * c + input_gate * candidate


    output_gate = sigmoid(wo * x + bo)


    h = output_gate * tanh(c)

    print("Step:", step + 1)
    print("Input:", x)
    print("Forget gate:", round(forget_gate,3))
    print("Input gate:", round(input_gate,3))
    print("Candidate value:", round(candidate,3))
    print("Cell state:", round(c,3))
    print("Hidden state:", round(h,3))
    print("-------------------")


prediction = 1.2 * h + 0.5

print("\nPredicted next value:", round(prediction,3))
```
